Import and call load_iris function

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris() # contains keys and values
```

```python
print("Keys of iris_dataset: \n{}".format(iris_dataset.keys()))
```

DESCR is a short description od the dataset

```python
print(iris_dataset['DESCR'][:193]+"\n...") # Beginning of the description
```

The value of the key target_names contains the species of flower that we want to predict (labels).

```python
print("Target bames: {}".format(iris_dataset['target_names']))
```

The value of feature_names contains the description of each feature (atributes or features).

```python
print("Feature names: \n{}".format(iris_dataset['feature_names']))
```

**data** contains the numerical values of the features.

```python
print("Type of data: {}".format(iris_dataset["data"]))
```

rows correspond to flowers and columns to feature measurements

```python
print("Shape of data: ", iris_dataset['data'].shape)
```

The data contains measurements for 150 different flowers (samples).

The feature values for the first five samples.

```python
print("First five rows of data:\n", iris_dataset['data'][:5])
```

The target array contains the species of each of the flowers that were measured.

```python
print("Type of target: ", type(iris_dataset['target']))
```

The species are encoded as integers from 0 to 2, where 0 means *setosa*, 1 means *versicolor* and 2 means *virginica*.

```python
print("Target:\n", iris_dataset['target'])
```

We will randomly split the data into two parts: training set (75%) and test set (25%).

Let's use  *train_test_split* to split the data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0) 
# X corresponds to the data, while y corresponds to the labels
```

*X_train* contains 75% of the dataset rows and *X_test* contains the remaining 25%.

```python
print("X_train shape: ", X_train.shape)
print("X_test shape: ", X_test.shape)
print("y_train shape: ", y_train.shape)
print("y_test shape: ", y_test.shape)
```

Create dataframe from data X_train

Label the columns using the strings in iris_dataset.feature_names

```python
import pandas as pd
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)
```

Create a scatter matrix from the dataframe, color by y_train

```python
import mglearn
pd.plotting.scatter_matrix(iris_dataframe, c=y_train, 
                           figsize=(15,15), marker='o', 
                           hist_kwds={'bins': 20}, s=60, 
                           alpha=.8, cmap=mglearn.cm3)
```

We can see that the three classes seem to be relatively well separated using the sepal and petal measurements. This means that a machine learning algorithm will likely be able to separate them.

We will use the K Neirest Neighbors classifier. The most important parameter in this algorithm is the number of neighbors, which we will set to 1.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=1)
```

To build the model on the training set, we call the *fit* method of knn object, which takes as arguments the numpy array X_train and y_train, containing all the information of the training. The *fit* method returns the knn object itself (and modifies it in place), so we get a string representation of our classifier.

```python
knn.fit(X_train, y_train)
```

We can now make predictions using the constructed model, with new data.

```python
import numpy as np
X_new = np.array([[5, 2.9, 1, 0.2]])
prediction = knn.predict(X_new)
print("Prediction: ", prediction)
print("Predicted target name: ", iris_dataset['target_names'][prediction])
```

Our model predicts that this new iris belongs to the class 0, meaning its species is *setosa*.

But, how do we know we can trust our model? We can make a prediction for each iris in the test data and compare it against its label. We can measure how well the model works by computing the *accuracy*, which is *the fraction of flowers for which the right species was predicted*.

```python
y_pred = knn.predict(X_test)
print("Test set predictions: ", y_pred)

# The score method of the knn object will compute the test set accuracy.
print("Test set core: ", np.mean(y_pred == y_test))
```

For this model, the test set accuracy is about 0.97, which means we made the right prediction for 97% of the irises in the test set. Under some mathematical assumptions, we can expect our model to be correct 97% of the times for new irises.
